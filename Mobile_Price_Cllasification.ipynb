Traing Data

```python
X=train_data[['battery_power', 'blue', 'clock_speed','dual_sim', 'fc', 'four_g', 'int_memory', 'n_cores', 'pc',
      'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'three_g', 'touch_screen', 'wifi']]
y=train_data['price_range']
```

Differntiating training data and testing data

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

Feeding data to the Model

```python
from sklearn.svm import SVC

model=SVC()
model.fit(X_train,y_train)
```

Prediction of trainig data

```python
predictions=model.predict(X_test)
print(predictions)
```

E valuting Model Performance

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predictions))
```
